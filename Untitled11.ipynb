```python
def waveshow(data, sr, emotion):
    plt.figure(figsize=(10,4))
    plt.title(emotion, size=20)
    librosa.display.waveshow(data, sr=sr)
    plt.show()
    
def spectogram(data, sr, emotion): 
    x = librosa.stft(data)
    xdb = librosa.amplitude_to_db(abs(x))
    plt.figure(figsize=(11,4))
    plt.title(emotion, size=20)
    librosa.display.specshow(xdb, sr=sr, x_axis='time', y_axis='hz')
    plt.colorbar()
    
```

```python
emotion = 'angry'
path = df['speech'][df['label']==emotion][1]
data, sampling_rate = librosa.load(path)
waveshow(data, sampling_rate, emotion)
spectogram(data, sampling_rate, emotion)
Audio(path)
```

```python
df=df.drop(df.index[0])
```

```python
##feature extraction
def extract_mfcc(filename):
     y, sr = librosa.load(filename, duration=3, offset=0.5)
     mfcc = np.mean(librosa.feature.mfcc(y=y, sr=sr, n_mfcc=40).T, axis=0)
     return mfcc
```

```python
extract_mfcc(df['speech'][1])
```

```python
X_mfcc = df['speech'].apply(lambda x: extract_mfcc(x))
```

```python
from sklearn.preprocessing import OneHotEncoder
enc = OneHotEncoder()
y = enc.fit_transform(df[['label']])
```

```python
y = y.toarray()
```

```python
from keras.models import Sequential
from keras.layers import Dense, LSTM, Dropout

model = Sequential([
    LSTM(256, return_sequences=False, input_shape=(40,1)),
    Dropout(0.2),
    Dense(128, activation='relu'),
    Dropout(0.2),
    Dense(64, activation='relu'),
    Dropout(0.2),
    Dense(7, activation='softmax')
])

model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
model.summary()
```

```python
X = [x for x in X_mfcc]
X = np.expand_dims(X, -1)
```

```python
# Train the model
history = model.fit(X, y, validation_split=0.2, epochs=50, batch_size=64)
```

```python
train_loss, train_accuracy = model.evaluate(X, y, verbose=0)
print(f"Training Accuracy: {train_accuracy:.4f}")
print(f"Training Loss: {train_loss:.4f}")
```

```python
val_loss_per_epoch = history.history['val_loss']
```

```python
for epoch, loss in enumerate(val_loss_per_epoch, start=1):
    print(f"Epoch {epoch}: Validation Loss = {loss:.4f}")
```

```python
final_val_loss = val_loss_per_epoch[-1]
print(f"Final Validation Loss: {final_val_loss:.4f}")
```
